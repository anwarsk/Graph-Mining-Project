```python
from sklearn.naive_bayes import BernoulliNB

def PU_Learner(input_data_train):
    #classifier = DecisionTreeClassifier(criterion='entropy', max_depth=5)
    #classifier = KNeighborsClassifier(n_neighbors=7)
    classifier = AdaBoostClassifier(n_estimators=10)
    #classifier = BernoulliNB()
    #classifier = MLPClassifier(hidden_layer_sizes=[70,70,70])
    isConverged = False
    P = input_data_train[input_data_train['Cited'] == 1]
    U = input_data_train[input_data_train['Cited'] != 1]
    
    U = U.reset_index(drop=True)
    RN = U
    RN_threshold = 0.95
    improve_factor = 0.95
    iterationCount = 0
    while(not isConverged):
        X_input = P.append(RN)
        X_input = X_input.reset_index(drop= True)
        y_input = X_input['Cited']
        X_input = X_input.drop(['Cited', 'AuthorId', 'ArticleId'], axis=1)
        #print("Input Size: ", X_input.size)
        classifier.fit(X_input, y_input)
        
        U_probabilities = classifier.predict_proba(U.drop(['Cited', 'AuthorId', 'ArticleId'], axis=1))
        #print(U_probabilities)
        RN_index = np.where(U_probabilities[:,0]>RN_threshold)[0]
    
        RN = U.iloc[RN_index,:]
       # print(RN.head())
        #print(U.head())
        U = U.drop(RN_index, axis=0)
        U = U.reset_index(drop=True)
        RN = RN.reset_index(drop=True)
        print("Iteration# %d \n  Input Size=%d \n  Number of RN=%d " % (iterationCount+1, X_input.size, len(RN)))
        iterationCount+=1
        RN_threshold *= improve_factor
        if(len(RN) <1):
            isConverged = True
            print("NO RN Found")
        if(iterationCount >30):
            isConverged = True
            print("Max Iteration Count Reached")
    return classifier
```
